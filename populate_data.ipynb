```python
import pandas as pd
import numpy as np

# Define the structure for the dummy dataset
learners = pd.DataFrame({
    'Learner_ID': range(1, 11),
    'Name': ['Learner' + str(i) for i in range(1, 11)],
    'Age': np.random.randint(18, 30, size=10),
    'Gender': np.random.choice(['Male', 'Female'], size=10),
    'Branch': np.random.choice(['Computer Science', 'Information Technology', 'Electronics'], size=10),
    'Qualification': np.random.choice(['Bachelor', 'Master', 'PhD'], size=10),
    'Background_Knowledge': np.random.choice(['Beginner', 'Intermediate', 'Advanced'], size=10),
    'Learning_Style': np.random.choice(['Active', 'Reflective', 'Sensing', 'Intuitive', 'Visual', 'Verbal', 'Sequential', 'Global'], size=10)
})

learning_materials = pd.DataFrame({
    'Material_ID': range(1, 21),
    'Title': ['Material' + str(i) for i in range(1, 21)],
    'Structure': np.random.choice(['Atomic', 'Collection', 'Networked', 'Hierarchical', 'Linear'], size=20),
    'Format': np.random.choice(['Text', 'Video'], size=20),
    'Interactivity_Type': np.random.choice(['Active', 'Expository', 'Mixed'], size=20),
    'Learning_Resource_Type': np.random.choice(['Exercise', 'Quiz', 'Slide', 'Document', 'Video'], size=20),
    'Interactivity_Level': np.random.choice(['Very Low', 'Low', 'Medium', 'High', 'Very High'], size=20),
    'Difficulty': np.random.choice(['Very Easy', 'Easy', 'Medium', 'Difficult', 'Very Difficult'], size=20)
})

learner_logs = pd.DataFrame({
    'Log_ID': range(1, 31),
    'Learner_ID': np.random.choice(learners['Learner_ID'], size=30),
    'Material_ID': np.random.choice(learning_materials['Material_ID'], size=30),
    'Time_Spent': np.random.randint(5, 120, size=30),  # Time spent in minutes
    'Rating': np.random.randint(1, 6, size=30)  # Rating out of 5
})



learners, learning_materials, learner_logs
```

```python
learners.to_csv('learners.csv'), learning_materials.to_csv('learning_materials.csv'), learner_logs.to_csv('learner_logs.csv')
```
